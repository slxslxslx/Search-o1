```python
import json

# ============ 第一步：读取 seal-0-0528.jsonl，构建 id -> Question 的映射 ============
seal_lookup = {}
with open('raw/seal-0-0528.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        record = json.loads(line)
        seal_lookup[record['id']] = record['Question']

# ============ 第二步：读取 all.jsonl，补充缺失的 Question 字段 ============
updated_records = []
missing_count = 0

with open('/root/autodl-tmp/project/Search-o1/outputs/runs.baselines/seal0.qwen3.5-9b.search_o1/all.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        record = json.loads(line)
        
        # 检查是否缺少 Question 字段
        if 'Question' not in record or not record['Question']:
            rid = record.get('id')
            if rid in seal_lookup:
                record['Question'] = seal_lookup[rid]
                missing_count += 1
            else:
                print(f"⚠ 警告: id={rid} 在 seal-0-0528.jsonl 中找不到对应记录！")
        
        updated_records.append(record)

# ============ 第三步：将更新后的数据写回 all.jsonl ============
with open('/root/autodl-tmp/project/Search-o1/outputs/runs.baselines/seal0.qwen3.5-9b.search_o1/all.jsonl', 'w', encoding='utf-8') as f:
    for record in updated_records:
        f.write(json.dumps(record, ensure_ascii=False) + '\n')

print(f"完成！共补充了 {missing_count} 条缺失的 Question 字段")
```
